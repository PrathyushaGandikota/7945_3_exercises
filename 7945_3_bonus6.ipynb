```python
df.drop_duplicates(keep=False,inplace=True) 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,5))
sns.distplot(df['gross'].dropna(),hist=True,bins=10)



```

```python
df1=df.groupby("title_year")["budget"].sum()
df2=df1.head()
df2
```

```python
df2=df.groupby(['title_year','language']).agg({'movie_title':pd.Series.count,'budget':np.sum}).reset_index()
df2.plot(x='title_year', y='budget', figsize=(8,5) , color='blue')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
ax = sns.violinplot(x="title_year", y="gross", data=df.head(10),split=True)
```
